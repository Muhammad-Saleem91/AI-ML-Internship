```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

# 1. Loading the .env file
load_dotenv()

# --- CONFIGURATION ---

API_KEY = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=API_KEY)
```

```python
# --- 2. THE PERSONA (System Prompt) ---
# We strictly define safety filters here 
system_instruction = """
You are a helpful, empathetic, and clear AI medical assistant.
Your goal is to provide general health information (e.g., hygiene, diet, sleep) and explain symptoms in simple terms.

CRITICAL SAFETY RULES:
1. You are NOT a doctor. Do NOT provide diagnoses, interpret lab results, or prescribe medication.
2. If a user asks about a life-threatening emergency (chest pain, trouble breathing, severe bleeding), IMMEDIATELY tell them to call emergency services.
3. Always end your response with: "\n\n(Disclaimer: I am an AI, not a doctor. Please consult a professional for medical advice.)"
"""
```

```python
# Initialize the model with the system instruction
# 'gemini-1.5-flash' is great for this as it's fast and free-tier eligible
model = genai.GenerativeModel(
    model_name="gemini-2.5-flash",
    system_instruction=system_instruction
)

# Start a chat session (keeps memory of the conversation)
chat_session = model.start_chat(history=[])
```

```python
# --- 3. THE CHAT LOOP ---
print("üè• HealthBot (Gemini Powered): Hello! I can answer general wellness questions. (Type 'quit' to exit)")

while True:
    user_input = input("\nYou: ")
    if user_input.lower() in ["quit", "exit"]:
        print("HealthBot: Stay healthy! Goodbye.")
        break
    
    try:
        # Send message to Gemini
        response = chat_session.send_message(user_input)
        print(f"HealthBot: {response.text}")
        
    except Exception as e:
        print(f"Error: {e}")
```
