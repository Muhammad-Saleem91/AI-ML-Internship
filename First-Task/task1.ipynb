Dataset loading and preprocessing 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
# Display the first 5 rows to verify data loading
print("\nFirst 5 rows:")
df.head()
```

```python
# Inspect the Data Structure
print("Dataset Shape:", df.shape)       
print("\nColumn Names:", df.columns)
```

```python
df.isna().sum() # No null values found as dataset is clean
```

```python
# Summary Statistics
print("\nDataset Info:")
print(df.info())            

print("\nStatistical Summary:")
print(df.describe())        
```

Data visualization and exploration 


```python
# Set the visual style
sns.set_style("whitegrid")

# 1. Histograms: Show value distributions [cite: 18]
# We use a histogram for each numerical feature
df.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle('Feature Distributions (Histograms)', fontsize=16)
plt.show()
```

```python
# 2. Box Plots: Identify outliers [cite: 19]
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title('Box Plots to Identify Outliers', fontsize=16)
plt.show()
```

```python
# 2. Box Plots: Identify outliers [cite: 19]
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title('Box Plots to Identify Outliers', fontsize=16)
plt.show()
sns.pairplot(df, hue='species', height=2.5)
plt.suptitle('Relationships Between Features (Scatter Plots)', y=1.02, fontsize=16)
plt.show()
```

Explanation of results and final insights:

The Scatter Plots revealed that the Setosa species is linearly separable from the others. The Histograms showed that Petal Length is bimodal, indicating a clear size difference between species. Finally, the Box Plots identified several outliers in the Sepal Width feature that may require further investigation.
